#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).


#these lines of code take the spread return for each of the datasets (early, middle and late) and divide them into the bitstamp_sliced dataset to turn them into percentage returns

#early
spread_return_early = arbitrage_spread_early.loc[arbitrage_spread_early > 0] / bitstamp_sliced["Close"].loc["2018-01-16"]
#middle
spread_return_middle = arbitrage_spread_middle.loc[arbitrage_spread_middle > 0] / bitstamp_sliced["Close"].loc["2018-02-24"]
#late
spread_return_late = arbitrage_spread_late.loc[arbitrage_spread_late > 0] / bitstamp_sliced["Close"].loc["2018-03-29"]

# Review the spread return DataFrame

#prints spread returns for each of the spread_return data sets, early, middle and late
print("Early")
display(spread_return_early.dropna())
print("")
print("Middle")
display(spread_return_middle.dropna())
print("")
print("Late")
display(spread_return_late.dropna())
```
